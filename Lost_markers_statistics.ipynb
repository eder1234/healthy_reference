Count the nans by movement

```python
list_M_val = [[],[],[],[],[]]
list_M = ['1', '2', '3', '4', '5']

for i in range(5):
    dataset_M = dataset.only_M(list_M[i])
    for j in range(len(dataset_M)):
        x_0 = dataset_M[j].markers[0].reshape(-1,3)[:,0]
        list_M_val[i].append(x_0)
```

```python
import numpy as np

n_nan = [[],[],[],[],[]]
for k in range(5):
    np_M = np.array(list_M_val[k])
    n_nan[k] = sum(np.isnan(np_M), 0)
    
```

```python
labels = dataset.masks[0].labels
```

```python
%matplotlib qt

import matplotlib.pyplot as plt

X_axis = np.arange(108)

plt.bar(X_axis - 0.4, n_nan[0], 0.1, label = 'M1')
plt.bar(X_axis - 0.2, n_nan[1], 0.1, label = 'M2')
plt.bar(X_axis      , n_nan[2], 0.1, label = 'M3')
plt.bar(X_axis + 0.2, n_nan[3], 0.1, label = 'M4')
plt.bar(X_axis + 0.4, n_nan[4], 0.1, label = 'M5')

plt.xticks(X_axis, labels)
plt.title("Number of nans per M and marker")
plt.legend()
plt.show()
```
