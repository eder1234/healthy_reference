```python
import visualization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
import functions
import pathlib
```

```python
with open('list_hr_mean_norm.pkl', 'rb') as file:
    list_hr_m = pickle.load(file)
```

```python
# Get the csv's filenames in the folder 
dataset_path = "/home/rodriguez/Documents/FaceMoCap_ML_Project/Data FaceMoCap/Sujets Pathologiques/VJ/20220315/"
ref_csv =  sorted(str(p) for p in pathlib.Path(dataset_path).glob("*.csv"))
```

```python
# Transform the patient's data following the same process as the healthy reference
# The csv's dataframes are organized here
list_ref = functions.create_list_ref(ref_csv)
list_dataset = functions.create_list_dataset(list_ref)

# The interpolation (Spline) and cleaning process are carried out here
list_ds_interpolated = functions.interpolate_list(list_dataset)
# The duration of each sample is set to 500 frames
list_ds_int_fix_dataset = functions.create_fixed_duration_dataset(list_ds_interpolated, fixed_duration=500)
# This line carries out the reference frame change transformation
list_ds_int_fix_dental = functions.dental_frame(list_ds_int_fix_dataset)
# If the displacement is selected as the feature of interest, use this function
list_ds_int_fix_dental_disp = functions.displacement_list(list_ds_int_fix_dental, norm = True)

list_VJ_m_03 = list_ds_int_fix_dental_disp
```

```python
%matplotlib inline

anomaly_mean_list_03 = []
for mov in range(5):
    mean_hr = np.nanmean(list_hr_m[mov], axis=0)[3:]
    mean_px = np.nanmean(list_VJ_m_03[mov][0], axis=0)[3:]
    anomaly = abs (mean_hr - mean_px)
    anomaly_mean_list_03.append(np.nanmean(anomaly))
    visualization.faciograph_px(mean_hr, mean_px)
```
